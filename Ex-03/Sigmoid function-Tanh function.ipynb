# Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random values
random_values = np.random.randn(100)

# Compute sigmoid values
sigmoid_values = 1 / (1 + np.exp(-random_values))

# Compute tanh values
tanh_values = np.tanh(random_values)

# Plot the results

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(random_values, sigmoid_values, color='blue', label='Sigmoid')
plt.subplot(1, 2, 2)
plt.scatter(random_values, tanh_values, color='red', label='Tanh')
plt.show()
```
