```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Generate random data
num_obs = 50
x1 = np.random.uniform(low=-10.0, high=10.0, size=num_obs)
noise = np.random.normal(loc=0.0, scale=5.0, size=num_obs)
y = 10 + 2 * x1 + noise
```

```python
# Plot the original data
plt.figure(figsize=(10, 7))
plt.plot(x1, y, 'o')
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Scatter plot of x1 vs y')
plt.show()
```

```python
# Reshape x1 for the linear regression model
x1_reshaped = x1.reshape(-1, 1)

# Fit the linear regression model
lr = LinearRegression().fit(x1_reshaped, y)
```

```python
# Plot the original data and the regression line
plt.figure(figsize=(10, 7))
plt.plot(x1, y, 'o')
x_chart = np.linspace(x1.min(), x1.max(), num=100).reshape(-1, 1)
plt.plot(x_chart, lr.intercept_ + lr.coef_[0] * x_chart, color='red')
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Scatter plot of x1 vs y with regression line')
plt.show()
```

```python
# Predict new values
x_new = np.array([0, 1, 2, 3]).reshape(-1, 1)
y_prediction = lr.predict(x_new)
print("Predicted values:", y_prediction)
```
