# PANDAS

```python
import pandas as pd  
import numpy as np  
#Creating Series from Array
arr = np.array(['P','a','n','d','a','s'])  
a = pd.Series(arr)  
print("Series from array:")
print(a) 
```

```python
#Series from dictionary
arr = {'x' : 0., 'y' : 1., 'z' : 2.}  
b = pd.Series(arr)  
print("\n\nSeries from dictionary:\n")
print (b)
```

```python
#Series using scalar
x = pd.Series(4, index=[0, 1, 2, 3])  
print("\nSeries using scalar\n")
print (x) 
```

```python
#Series through index 
x = pd.Series([1,2,3],index = ['a','b','c'])  
print("\nSeries through index:") 
print (x)
```

```python
a=pd.Series(data=[1,2,3,4])   
print("\n a series:\n",a)
```

```python
print("\nIndex: \n",a.index)
```

```python
print("\nValues: \n",a.values)
```

```python
print("\nShape: ",a.shape)   
```

```python
print("\nDimension: ",a.ndim)   
```

```python
print("\nSize: ",a.size) 
```

```python
emp = pd.Series(['Parker', 'John', 'Smith', 'William'])   
id =pd.Series([102, 107, 109, 114])     
frame = { 'Emp': emp, 'ID': id }   
result = pd.DataFrame(frame) 
print("\nSeries to Data frame\n")
print(result)
```

```python
 left = pd.DataFrame({  
   'id':[1,2,3,4,5],  
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],  
   'sub':['sub1','sub2','sub4','sub6','sub5']})  
right = pd.DataFrame({  
    'id':[1,2,3,4,5],  
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],  
   'sub':['sub2','sub4','sub3','sub6','sub5']})  
print(left)
```

```python
print(right)
```

```python
print(pd.merge(left,right,on='id'))
```

# MATPLOTLIB

```python
import matplotlib.pyplot as plt
xpoints = np.array([0,1,2,3,4,5])
ypoints = np.array([0,20,25,32,45,50])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
plt.plot(xpoints, ypoints,marker= 'o')  
plt.xlabel("index")
plt.ylabel("value")
plt.show()
```

```python
 y = np.array([35, 25, 25, 15])
 mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
 myexplode = [0.2, 0, 0, 0]
 plt.pie(y, labels = mylabels, explode = myexplode)
 plt.show()
```

```python
 y = np.array([35, 25, 25, 15])
 mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
 plt.pie(y, labels = mylabels)
 plt.legend(title = "Four Fruits:")
 plt.show() 
```
